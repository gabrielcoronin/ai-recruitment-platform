### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Load data

```python
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv("../data/processed/df.csv")
```

### EDA - Decision Recruitment Dataset

Now, let's start the exploratory analysis, focusing on data patterns that can be used to optimize the recruitment process.

#### Overview

```python
print("Dataset shape:", df.shape)
df.head()
```

#### Distribution of Contract Types in Vacancies



Understanding the types of contract can be a key point in analyzing how vacancies are offered and the type of talent we are looking for.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
df['vacancy_contract_type'].value_counts().nlargest(5).plot(kind='bar', color='skyblue')
plt.title('Distribution of Contract Types in Vacancies')
plt.xlabel('Type of Contract')
plt.ylabel('Number of vacancies')
plt.xticks(rotation=45)
plt.show()
```

PJ is the most common, so perhaps companies are prioritizing flexibility and fewer labor charges.

#### Distribution of Education Level

Education is an important factor in ensuring that candidates have the necessary training for the vacancies.

```python
plt.figure(figsize=(10,6))
df['vacancy_education_level'].value_counts().nlargest(5).plot(kind='bar', color='lightgreen')
plt.title('Distribution of Education Level')
plt.xlabel('Level of Education')
plt.ylabel('Number of vacancies')
plt.xticks(rotation=45)
plt.show()
```

Most of the vacancies require full higher education, showing the importance of university.

#### Candidate certifications

Certifications are indicative of technical skills and can directly impact the likelihood of a candidate being successful in their application.

```python
plt.figure(figsize=(10,6))
df['candidate_certifications'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution of Certifications')
plt.xlabel('Number of Certifications')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

Most students are not certified

#### Distribution of Candidates' English Levels


English is a crucial skill for many vacancies in the IT sector. Let's take a look at how candidates' level of English aligns with the vacancies available.

```python
plt.figure(figsize=(10,6))
df['candidate_english_level'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Distribution of Candidates\' English Level')
plt.xlabel('English Level')
plt.ylabel('Number of Candidates')
plt.xticks(rotation=45)
plt.show()
```

Most applicants don't fill in their level of English, so if applicants know that English proficiency is important, but don't fill in this information, it may suggest that they are not confident in their level of English and prefer not to declare it.

#### Distribution of Candidates per Vacancy


This analysis helps to understand whether certain vacancies are attracting more candidates than others, which may indicate greater interest or need in the market.

```python
# Candidate distribution by job vacancy region (top 5)
vacancy_counts = df['vacancy_region'].value_counts().nlargest(5)
vacancy_counts.plot(kind='bar', figsize=(12,6), color='orange')
plt.title('Candidates Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Candidates')
plt.xticks(rotation=45)
plt.show()
```

The highest concentration of applicants is in São Paulo, followed by Minas Gerais, Ceará, Rio de Janeiro and Pernambuco, which indicates that these states have the greatest interest in vacancies, with São Paulo leading the way, while other regions may need specific strategies to increase participation.
